```python
def magic_fomular_by_pd(path) :


    # Import pandas module and read excel workbook
    import pandas as pd
    file_path = path
    #C:\\Users\\jongh\\Downloads\\주식 책 데이터\\마법공식 데이터.xlsx
    # Read PER sheet from workbook and filtering data over than 0 and sorting the data as ascending. Add Rank column and assgin rank value.
    per_data = pd.read_excel(file_path, sheet_name = 'PER', index_col = 0)
    filtered_per = per_data[per_data['PER'] > 0]
    sorted_per = filtered_per.sort_values(by = 'PER')
    sorted_per['PER Rank'] = sorted_per['PER'].rank()
    
    
    # Read ROA sheet from workbook and filtering data except empty value. And modify column name to 'ROA', sorting data as descending. Add Rank column and assign rank value
    roa_data = pd.read_excel(file_path, sheet_name = 'ROA', index_col = 0)
    filtered_roa = roa_data.dropna()
    filtered_roa.columns = ['ROA']
    sorted_roa = filtered_roa.sort_values(by = 'ROA', ascending = False)
    sorted_roa['ROA Rank'] = sorted_roa['ROA'].rank(ascending = False)
    
    # Joinning the two data frames into one data frame. 
    total_df = pd.merge(sorted_per, sorted_roa, how = 'inner', left_index = True, right_index = True)
    
    # Add total rank column and sum ROA rank and PER rank in total rank column, assign rank value in each company
    total_df['Total Rank'] = (total_df['PER Rank'] + total_df['ROA Rank'])
    total_rank = total_df.sort_values(by = 'Total Rank')
    total_rank['Total Rank'] = total_rank['Total Rank'].rank()

    return total_rank

magic_fomular_by_pd('C:\\Users\\jongh\\Downloads\\주식 책 데이터\\마법공식 데이터.xlsx')
```
