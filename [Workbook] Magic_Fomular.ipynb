```python
# openpyxl module import and load excel workbook and worksheet
import openpyxl
load_wb = openpyxl.load_workbook('C:\\Users\\jongh\\Downloads\\주식 책 데이터\\마법공식 데이터.xlsx')
load_ws = load_wb['PER']

#'PER' sheet load and create PER dictionary, if PER value is more than 0 then load the value.
per_dict = {}

for row in load_ws.iter_rows(min_row = 2, max_row = load_ws.max_row, min_col = 1, max_col = load_ws.max_column) :
    data = [cell.value for cell in row]
    name = data[0]
    value = data[1]
    if value > 0 :
        per_dict[name] = value

# operator module import and sorting PER data in sorted_per variable
import operator
sorted_per = sorted(per_dict.items(), key = operator.itemgetter(1))

# assign PER score ascending
per_rank = {}

for num, firm in enumerate(sorted_per) : 
    per_rank[firm[0]] = num + 1

#load ROA sheet from excel workbook and load and if ROA value is not none and over than 0, save ROA values in roa_dict dictionary
load_ws2 = load_wb['ROA']
roa_dict = {}

for row in load_ws2.iter_rows(min_row = 2, max_row = load_ws2.max_row, min_col = 1, max_col = load_ws2.max_column) :
    data = [cell.value for cell in row]
    name = data[0]
    value = data[1]
    if value is not None and value > 0 :
        roa_dict[name] = value

# sorting ROA data in sorted_roa variable in descending
sorted_roa = sorted(roa_dict.items(), key = operator.itemgetter(1), reverse = True)

# assign ROA score in ascending
roa_rank = {}

for num, firm in enumerate(sorted_roa) :
    roa_rank[firm[0]] = num + 1

# create total_rank dictionary and assign values in sum of the roa rank and per rank
total_rank = {}

for name in roa_rank.keys() :
    if name in per_rank.keys() :
        total_rank[name] = per_rank[name] + roa_rank[name]

# sorting values in total_rank dictionary in ascending and assign values in sorted_total_rank variable, assign rank score in magic_rank dictionary
sorted_total_rank = sorted(total_rank.items(), key = operator.itemgetter(1))

magic_rank = {}

for num, firm in enumerate(sorted_total_rank) : 
    magic_rank[firm[0]] = num + 1

# print result
print(magic_rank)
```
